### 01 두 직선의 교차점 구하기

```python
def simultaneous_eq(a,b):
    if a[0] == 0:
        y = Fraction(a[2], a[1])
        x = Fraction(b[2] - b[1]*y, b[0])
    elif a[1] == 0:
        x = Fraction(a[2], a[0])
        y = Fraction(b[2] - b[0]*x, b[1])
    elif b[0] == 0:
        y = Fraction(b[2], b[1])
        x = Fraction(a[2] - a[1]*y, a[0])
    elif b[1] == 0:
        x = Fraction(b[2], b[0])
        y = Fraction(a[2] - a[0]*x, a[1])
    else:
        a = np.array(a)
        b = np.array(b)
        c = b* Fraction(a[0],b[0])
        d = a - c
        y = Fraction(d[2],d[1])
        x = Fraction(a[2]-a[1]*y,a[0])
    return x, y
```

#### Question 4.18

(2,10), (-3, -5)를 지나는 직선은 y = 3x + 4이고 이 직선이 위의 두 직선과 만난다.

y = 3x + 4와 y-x-1 = 0의 교점은 (-3/2, -1/2) 이다. (아래 코딩참조)

```python
def simultaneous_eq(a,b):
    if a[0] == 0:
        y = Fraction(a[2], a[1])
        x = Fraction(b[2] - b[1]*y, b[0])
    elif a[1] == 0:
        x = Fraction(a[2], a[0])
        y = Fraction(b[2] - b[0]*x, b[1])
    elif b[0] == 0:
        y = Fraction(b[2], b[1])
        x = Fraction(a[2] - a[1]*y, a[0])
    elif b[1] == 0:
        x = Fraction(b[2], b[0])
        y = Fraction(a[2] - a[0]*x, a[1])
    else:
        a = np.array(a)
        b = np.array(b)
        c = b* Fraction(a[0],b[0])
        d = a - c
        y = Fraction(d[2],d[1])
        x = Fraction(a[2]-a[1]*y,a[0])
    return x, y
```

#### Question 4.20

우선 독자수에 대한 점수를 매기는 코딩함수를 만들겠습니다.

```python
def Audience(n):
    if n < 10:
        score = 0
    else:
        score = (n-10)*0.5
    return score
```

```python
def Star(star):
    if star <= 3:
        score = -1
    elif star <= 6:
        score = 0
    elif star <= 9:
        score = 1
    else:
        score = 2
    return score
```

```python
audience = np.arange(0, 30, 1)
star = np.arange(0,10,1)

score_aud = np.zeros(len(audience))
score_star = np.zeros(len(star))

for i in range(len(audience)):
    score_aud[i] = Audience(audience[i])
    
for j in range(len(star)):
    score_star[j] = Star(star[j])

plt.plot(audience, score_aud)
plt.figure()
plt.plot(star, score_star)
```

```python
score = np.zeros((len(audience), len(star)))
for i in range(len(audience)):
    for j in range(len(star)):
        score[i,j] = Audience(audience[i]) + Star(star[j])
score.shape
```

```python
print ('가능한 최고 점수는 {} 이다.'.format(np.max(score)))
```

#### Question 4.22

```python
def saving(money, years, rate):
    result = money*(1+rate/100)**years
    return result
```

만원을 3%, 5%, 7%이율로 예금하였을때 10년이 될때까지 매년 총 돈의 액수를 함수로 그리시오.

```python
money = 10000
years = np.arange(0,11)
rate = [3, 5, 7]
total_money = np.zeros((len(rate), len(years)))
for i in range(len(rate)):
    for j in range(len(years)):
        total_money[i,j] = saving(money, years[j], rate[i])
    plt.plot(years, total_money[i,:], label = '{}%'.format(rate[i]))
plt.legend(fontsize = 14)
plt.grid()
plt.xlabel('Years', fontsize = 14)
plt.ylabel('Total money', fontsize = 14)
```

위의 3,5,7% 이율에서 돈이 처음에 넣은돈의 2배가 될때까지 얼마나 기다려야 하는지 함수를 그려서 예측해 보시오.

10년이 아니라 더 오래 넣고 기다려야 됩니다.

```python
money = 10000
years = np.arange(0,25)
rate = [3, 5, 7]
total_money = np.zeros((len(rate), len(years)))
for i in range(len(rate)):
    for j in range(len(years)):
        total_money[i,j] = saving(money, years[j], rate[i])
    plt.plot(years, total_money[i,:], label = '{}%'.format(rate[i]))
plt.axhline(y = 20000, color = 'red', label = '20000')
plt.legend(fontsize = 14)
plt.grid()
plt.xlabel('Years', fontsize = 14)
plt.ylabel('Total money', fontsize = 14)
```

위 그래프를 보건데 3% 일때는 24년정도, 5%일때는 15년, 7%일때는 10년정도 걸립니다.

대략적으로 원금의 2배가 되기까지는 

$ 72 \div 이자율$

로 예측할 수 있습니다.

그렇담 부등호를 써서 정확히 몇년이상이 되어야 처음넣은 금액의 2배이상이 되는지 계산하시오.

```python
for i, val in enumerate(total_money[0,:]):
    if val > 20000:
        print ('{}% rate takes {} years'.format(3, i))
        break
```

```python
for i, val in enumerate(total_money[1,:]):
    if val > 20000:
        print ('{}% rate takes {} years'.format(5, i))
        break
```

```python
for i, val in enumerate(total_money[2,:]):
    if val > 20000:
        print ('{}% rate takes {} years'.format(7, i))
        break
```

#### break는 반복문 (for, while)을 그만두라는 명령어 입니다. 

즉 돈이 원금의 2배인 20000원이 되면 더이상 for문을 실행하지않고 빠져나오라는 의미입니다.
